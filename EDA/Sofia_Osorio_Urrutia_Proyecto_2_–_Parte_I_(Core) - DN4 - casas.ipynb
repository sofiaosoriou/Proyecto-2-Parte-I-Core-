## 2.Análisis Exploratorio de Datos (EDA) Inicial:

### Data Set N°4: Casas

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
df_ho = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df_ho.info()
```

```python
df_ho[df_ho.duplicated()]
```

```python
df_ho.describe().T
```

#### Visualizaciones

```python
plt.figure(figsize=(8, 5))
plt.hist(df_ho["MedInc"], bins=30, edgecolor='k', color='skyblue')
plt.title("Distribución de Mediana de Ingresos (MedInc)")
plt.xlabel("Mediana de Ingresos")
plt.ylabel("Frecuencia")
plt.show()
```

muestra cómo se distribuye el ingreso medio de los hogares.

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    df_ho["MedInc"], df_ho["HouseAge"],
    c=df_ho["AveRooms"], cmap="viridis", alpha=0.6
)
plt.colorbar(scatter, label="Promedio de Habitaciones (AveRooms)")
plt.title("Relación entre Mediana de Ingresos y Antigüedad de Casas")
plt.xlabel("MedInc")
plt.ylabel("HouseAge")
plt.show()
```

Relaciona ingresos y antigüedad de viviendas, y añade información de cuántas habitaciones promedio hay.

## Diagnóstico y Selección de Problema

Dataset N°4 : Casas ---> target: MedHouseVal (numerica) ---> Regressor  

Dado que el objetivo es predecir este valor en función de variables como ingresos, antigüedad de las viviendas o localización, el problema se percibe  como uno de regresión. Un modelo de regresión es adecuado porque permite cuantificar y predecir relaciones entre múltiples variables independientes y una variable dependiente continua, generando estimaciones precisas del precio promedio de la vivienda con características similares.
